Question 3 

 - Vérifions que l'équation $\dot{X} = f(X)$ admette des solutions lorsqu'on l'initialise dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$. Comme $f$ est continue sur $\mathbb{R} \times \mathbb{R}^2$, par le $théoreme \ de \ Péano - Arzelà$, pour tout $(t_0, x_0) \in \mathbb{R} \times \mathbb{R}^n$  il existe un intervalle I de $\mathbb{R}$ et une fonction $x : I \to \mathbb{R}^2$ tq $x \in S_{f}(t_0, x_0)$

- Comme f est même continûement différentiable par rapport à x, on est dans les conditions du théorème de Cauchy Lipschitz


Enfin, supposons des conditions initiales $(t_0,x_0=(x_{10},x_{20})) \in \mathbb{R} \times (\mathbb{R_{>0}})^2$

Par le théoreme de Cauchy Lipstchitz, il existe une unique fonction maximale $x$ solution du problème telle que $x(t_0) = x_0$ 

Par l'absurde, supposons qu'il existe $t_1$ dans le domaine d'existence de $x$ tq $x(t1) = (0, x_2(t_1))$ où $x_2(t_1)) \ne 0$

donc $x \in S_f(t1, x(t_1))$

Or on remarque que $\tilde{x}$ : $\begin{array}{ccccc}
 \mathbb{R} & \to & \mathbb{R} \\
 t & \mapsto & (0, x(t1)\ e^{\alpha (t-t_1)}) \\
\end{array} \in S_f(t1, x(t_1))$ et $\tilde{x}$ est définie sur $\mathbb{R}$, c'est une solution maximale. Ainsi par le théorème de Cauchy Lipschitz, $x = \tilde{x}$. Or $\tilde{x_1}(t_0) = 0 \ne x_1(t_0)$

Ce resultat est donc absurde : il est impossible que la coordonnée $x_1$ s'annule sur son domaine si elle est initialisée dans $\mathbb{R}^*_+$. Et comme elle est continue, elle ne peut pas non plus être négative par un corollaire du théorème des valeurs intermediaires. Ainsi, $x_1$ reste dans $\mathbb{R_{>0}}$ si elle a été initialisée dans ce même domaine. Le raisonnement est exactement le même pour $x_2$. Enfin, le vecteur $x$ ne peut totalement s'annuler car la fonction nulle est la seule solution passant par (0,0).

Question 4 :

On note $I$ le domaine de définition de la solution  $x = (x_1,x_2)$. Comme $x$ est à image dans $\mathbb{R_{>0}} \times \mathbb{R_{>0}}$ , la fonction $t \to H(x_1(t),x_2(t))$ est definie sur $I$. Comme $x$ est $C^1$ sur $I$, $H$ l'est aussi et :
$\forall t \in I$ :
$H'(t) = \delta \dot{x_1}(t) - \gamma \frac{\dot{x_1}(t)}{x_1(t)} + \beta \dot{x_2(t)} - \alpha \frac{\dot{x_2}(t)}{x_2(t)} = 0$
donc $H$ est constante sur $I$. Notons $A$ cette constante.

on a donc : 

$A = \delta x_1 - \gamma \ln(x_1) + \beta x_2 - \alpha \ln(x_2)$

Montrons que $||x||$ est bornée, ce qui permettra de conclure, avec le théoreme du domaine maximal d'existence que $x$ est forcement définie sur $\mathbb{R}$.

Supposons que $||x|| \to +\infty$ . comme $x$ est à valeur dans $\mathbb{R_{>0}} \times \mathbb{R_{>0}}$ on a forcement $x_1$ et/ou $x_2 \to +\infty$.

or $\ln(y) = o(y)$ en $y \to+ \infty$ donc si $x_1$ diverge, pour que $A$ soit constante, il faut que $\beta x_2 -\alpha ln(x_2) \to - \infty$ ce qui est impossible car $\beta, \delta > 0$. $||x||$ est donc bornée.

Ainsi, la solution maximale d'un problème de Cauchy initialisé sur $\mathbb{R_{>0}} \times \mathbb{R_{>0}}$ est définie sur $\mathbb{R}$

### Question 6 :

```python
def solve_euler_explicit(f, x0, dt, t0, tf):
    """f est la fonction de la dérivée, x0 la valeur en t0, tf le temps final"""
    
    i = int((tf-t0)/dt)
    T = [t0] 
    X = [x0]
    
    for j in range(1,i) :
        #tous les calculs sont possibles avec des array
        X.append(X[-1]+dt*f(T[-1]+dt,X[-1]))
        T.append(T[-1]+dt)
    return T, X

#On test la fonction avec une fonction exponentielle, soit avec f(x)=x

def f2(t,x):
    return np.array(x)

T,X = solve_euler_explicit(f2, np.array([1]),0.01,0,5)
Y = np.exp(T)

#on trace les résultats

plt.plot(T,X,color='b',label = "Euler Explicite")
plt.plot(T,Y,color='g', label = "Courbe réelle")
plt.legend()
plt.title("Comparaison méthode implicite et courbe réelle \n -Méthode explicite-")
plt.plot()
```

On cherche à obtenir une estimation de l'erreur avec le modèle réel. Pour cela, on définit une fonction erreur qui prend en argument le pas de temps dt et qui retourne la valeur maximale de l'écart entre le modèle calculé et la valeur réelle.

```python
def erreur(dt) :
    T,X = solve_euler_explicit(f2, np.array([1]),dt,0,5)
    Y = np.exp(T)
    l=[]
    for i in range(0,len(X)) :
        l.append(abs(X[i]-Y[i]))
    k = float(max(l))
    return k
erreur=np.vectorize(erreur)
```

```python
DT=np.linspace(0.0001,0.1,1000)
ERR = [erreur(dt) for dt in DT]
plt.plot(np.log(DT),np.log(ERR),color='b',label='Erreur')
plt.plot(np.log(DT),np.log(DT), color='r', label='Fonction identité')
plt.legend()
plt.title("Erreur en fonction de dt \n -Echelle logarithmique-")
plt.plot()
```

On a tracé ici en échelle logarithmique la valeur maximale de l'erreur entre les valeurs obtenues à l'aide de l'algorithme et les valeurs réelles, et ce en fonction du pas de temps. Or on constate que l'erreur a la même pente que la foncton identité. Donc la fonction qui au logarithme du pas de temps associe le logarithme de l'erreur est une fonction affine de pente 1. Ainsi : $\exists a \in \mathbb{R}, \ np.log(ERR)=np.log(DT)+a \ \implies \exists k \in \mathbb{R}, ERR=k*DT$

Visuellement on constate alors que l'ordre de convergence est de 1.

De là, on en déduit que l'erreur tend vers 0 lorsque que le pas de temps tend également vers 0

```python
def solve_euler_implicit(f, x0, dt, t0, tf, itermax = 100):
    """f est la fonction de la dérivée, x0 la valeur en t0, tf le temps final"""
   
    eps = 10**(-6)
    
    T,X = solve_euler_explicit(f, x0, dt, t0, tf)
    
    #On définit la norme 2
    N = lambda X : sum(e**2 for e in X)**(0.5)
    
    for i in range(1,len(X)) :
        t, x0 = T[i], X[i] 
        
        #On définit F, et on cherche x1, point fixe de F
        F = lambda u : X[i-1] + dt*f(t,u)
       
        x = x0
        x1 = F(x0)
        
        reussi = False
        
        for k in range(1,itermax) :
            if N((x1-x)/x0)<eps :
                X[i]=x1
                reussi = True
                break
            x, x1 = x1, F(x1)
            
        if not reussi :
            raise ValueError("No convergence in {} steps".format(itermax))
    return T,X
    
```

### Question 11

Pour assurer la stabilité de $H$, il faut modifier le schéma d'Euler en ne considerant plus la fonction f mais la nouvelle fonction g (de la question précèdente) car cette derniere permet une convergence exponentielle de $H  \circ x$ vers $H_0$.


$k$ controle la vitesse de convergence de $H$ vers $H_0$. Au premier abord, plus $k$ est grand plus la vitesse de convergence semble être élevée. Cependant, on ne peut pas le prendre trop grand non plus. En effet en faisant un DL de $H(x^{j+1}) - H_0$ au premier ordre on obtient : 

$H(x^{j+1}) - H_0 = H(x^{j}+\Delta t_j.g(t_{j+1},x^{j+1}) - H_0 \approx H(x^{j+1}) - H_0 + \Delta t_j. ^t \nabla H(x^{j+1}) .g(t_j+1, x^{j+1})$

d'où : 
$H(x^{j+1}) - H_0 \approx H(x^{j}) - H_0 - k \Delta t_j ||\nabla H(x_j)||^2$


Ici, on se rend compte que si k est trop grand, l'élément différentiel $\Delta x = -k. \Delta t \nabla H (x)$ utilisé dans la méthode d'Euler pour calculer $x$ n'est plus négligeable et diminue la précision de la méthode.

```python
#On définit la fonction g introduite précédemment
x0 = np.array([1,2])
def g(t,X) :
    k=1
    x,y=X[0],X[1]
    grad_H = np.array([d-c/x,a-b/y])
    u_x = k*grad_H
    return f(t,X)-(H(x,y)-H(x0[0],x0[1]))*u_x
```

```python
T,X1 = solve_euler_implicit(g,np.array([1,2]),0.001,0,50)
plt.plot(T,[e[0] for e in X1], color = 'b', label = 'x1')
plt.plot(T,[e[1] for e in X1], color = 'orange', label = 'x2')
plt.title('Population de x1 et x2 en fonction du temps \n -Méthode implicite, fonction modifiée-')
plt.legend()
plt.plot()
```

```python
T,X1 = solve_euler_implicit(g,np.array([1,2]),0.001,0,200)
plt.plot(T,[e[0] for e in X1], color = 'b', label = 'x1')
plt.plot(T,[e[1] for e in X1], color = 'orange', label = 'x2')
plt.title('Population de x1 et x2 en fonction du temps \n -Méthode implicite, fonction modifiée, temps long-')
plt.legend()
plt.plot()
```

```python
Y = []
for i in range(0,len(X1)):
    Y.append(h(X1[i][0],X1[i][1]))
plt.plot(T,Y)
plt.title('Evolution de H en fonction du temps \n -Méthode implicite, fonction modifiée, temps long-')
plt.xlabel('Temps')
plt.plot()
```
